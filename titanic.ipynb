```python
# importing relevant modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# opening the dataset
titanic_df = pd.read_csv("titanic.csv")

# checking out the shape of the dataset
titanic_df.shape
```

```python
# removing unwanted columns - Name, Cabin, Ticket
titanic_df.drop(columns=["Name", "Cabin","Ticket"])
```

```python
# checking for missing values
missing_values_count = titanic_df.isnull().sum()
missing_values_count[0:12]
```

```python
# descriptive statistics of the age of passengers
titanic_df["Age"].describe()
```

Age data observations:
- The average age on the Titanic was 29(sd=14.5).
- The median age was 28.
- The youngest age on the ship was 4 months.
- The oldest age on the ship was 80.

```python
# count for how many survived
titanic_df.Survived.value_counts(normalize=True)
```

- The majority of the passengers did not survive the Titanic.
- The data shows that only 38% of the passengers survived.

```python
# Univariate Analysis
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
sns.histplot(titanic_df["Survived"],ax=axes[0,0])
sns.histplot(titanic_df["Sex"],ax=axes[0,1])
sns.histplot(titanic_df["Pclass"],ax=axes[0,2])
sns.histplot(titanic_df["SibSp"],ax=axes[1,0])
sns.histplot(titanic_df["Embarked"],ax=axes[1,1])
sns.histplot(titanic_df['Age'].dropna(),kde=True,ax=axes[1,2])
```

Observations:
- There were more males than females aboard the ship. 
- The majority of passengers were lowerclass, having purchased a third class ticket.
- Most of the passengers were betweeen the ages of 20 and 40.
- Majority of the passangers embarked from Southampton. 

```python
# Bivariate Analysis
figbi, axesbi = plt.subplots(2, 3, figsize=(18, 12))
titanic_df.groupby("Sex")['Survived'].mean().plot(kind='bar', ax=axesbi[0,0])
titanic_df.groupby("Pclass")['Survived'].mean().plot(kind='bar',ax=axesbi[0,1])
titanic_df.groupby("SibSp")['Survived'].mean().plot(kind='bar',ax=axesbi[0,2])
titanic_df.groupby("Parch")['Survived'].mean().plot(kind='bar',ax=axesbi[1,0])
titanic_df.groupby("Embarked")['Survived'].mean().plot(kind='bar',ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Age", data=titanic_df,ax=axesbi[1,2])
```

Observations:
- The most important factor in determining survival during the Titanic incident is Sex, as it shows that the survival rate of females was approx 75% compared to males which is approx 20%.
- Upper class passengers were definately show preference on lifeboats as those in first class had the highest survival rate (approx 65%).
- Women and Children were given preference on lifeboats too. The data shows that female had a higher survival rate. The age boxplot shows that the younger passengers survived more compared to older passengers.
- What is also interesting is the siblings survival rate. As the number of siblings increase, the survival rate decreases. This suggests that on the Titanic, larger families tended to stick together on the ship in hope of rescue.
